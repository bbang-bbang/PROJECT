# 5. 학습시키기

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import MinMaxScaler, StandardScaler


train_columns = [
 'Elevation',
 'Aspect',
 'Slope',
 'Horizontal_Distance_To_Hydrology',
 'Vertical_Distance_To_Hydrology',
 'Diagonally_Distance_To_Hydrology',   
 'Horizontal_Distance_To_Roadways',
 'Hillshade_9am',
 'Hillshade_Noon',
 'Hillshade_3pm',
 'Horizontal_Distance_To_Fire_Points',
 'Soil_Type',
 'Wilderness_Area']

data = train[train_columns]

imputer = SimpleImputer(strategy='mean')
data = imputer.fit_transform(data)

scaler = StandardScaler()
data = scaler.fit_transform(data)

train[train_columns] = data
```

```python
X_data = data
Y_data = train["Cover_Type"].values

print(np.unique(Y_data))
Y_data = Y_data - 1
print(np.unique(Y_data))

cover_type_to_label = {1:0, 2:1, 3:2, 4:3}
train["Cover_Type"] = train["Cover_Type"].map(lambda x: cover_type_to_label[x])

print(np.unique(Y_data))

from sklearn.model_selection import train_test_split

X_train, X_valid, Y_train, Y_valid = train_test_split(X_data, Y_data, test_size=0.2, stratify=Y_data)
```

```python
import lightgbm as lgb

lgb_train = lgb.Dataset(X_train, Y_train)
lgb_eval = lgb.Dataset(X_valid, Y_valid, reference = lgb_train)

params = {
     'boosting_type' : 'gbdt',
    'objective': 'multiclass',
    'metric': ['multi_logloss'],
    'num_classes': 4,
    'learning_rate': 0.2,
    'feature_fraction': 0.6,
    'bagging_fraction': 0.8,
    'bagging_freq': 10,
    'max_depth': 10,
    'verbosity': 1
}

lgb_model = lgb.train(params,
          lgb_train,
          num_boost_round = 100,
          valid_sets = lgb_eval,
          valid_names = ('validation')
          )
```
