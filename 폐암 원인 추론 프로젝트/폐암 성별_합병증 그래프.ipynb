```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
### 고혈압
```

```python
hypertension= pd.read_csv("폐암 합병증/lung_hypertension.csv")
hypertension
```

```python
hypertension.columns=['center','year', 'age', 'sex', 'hypertension','ncsNmvl','wholNcsDnmvl','ptCntNmvl','wholPtCntDnmvl']
hypertension
```

```python
sns.countplot(data=hypertension, x='hypertension', hue='sex',order=['Y','N'])
```

```python
hypertension_m=hypertension.loc[hypertension.sex=='M']
hypertension_f=hypertension.loc[hypertension.sex=='F']
```

```python
sns.countplot(data=hypertension_m, x='hypertension',order=['Y','N'])
```

```python
sns.countplot(data=hypertension_f, x='hypertension',order=['Y','N'])
```
